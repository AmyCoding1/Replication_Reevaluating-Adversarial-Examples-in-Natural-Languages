```python
import tqdm
tqdm.tqdm.pandas()

import pandas as pd
MR_CSV_PATH = '/content/drive/MyDrive/Project_AI/Replication-NLP-Adversarial-Examples-master/section_3_case_study/grammaticality/case_study_grammaticality_entailment/log_textfooler_bert_movie review.csv'
mr_df = pd.read_csv(MR_CSV_PATH)
mr_df.head()
```

```python
import language_tool_python
lang_tool = language_tool_python.LanguageTool("en-US")
```

```python
def count_errors(text): return len(lang_tool.check(text))
```

```python
mr_df['p1_errors'] = mr_df['original_text'].progress_map(count_errors)
mr_df['p2_errors'] = mr_df['perturbed_text'].progress_map(count_errors)
```

```python
(mr_df['p2_errors'] > mr_df['p1_errors']).mean()
```
